# **Tokyo 2020 Olympics Medals Analysis**

## Data Ingestion and Preparation

```python
import csv
```

```python
filename = "Tokyo Medals 2020.csv"
```

```python
medal_data = []
with open(filename, mode='r') as file:
    reader = csv.DictReader(file)
    for row in reader:

        country_info = {
            "Country": row["Country"],
            "Gold": int(row["Gold Medal"]),
            "Silver": int(row["Silver Medal"]),
            "Bronze": int(row["Bronze Medal"]),
            "Total": int(row["Total"]),
            "Rank": int(row["Rank By Total"])
        }
        medal_data.append(country_info)
```

## Displaying original Datasets

```python
for country in medal_data:
    print(country)
```

## Classification of Countries in Gold, Silver, Bronze, Balanced and Mixed Dominant Category

```python

Countries = {
    "Gold Dominant Category": [],
    "Silver Dominant Category": [],
    "Bronze Dominant Category": [],
    "Balanced Category": [],
    "Mixed Category": []
}

for country in medal_data:
    total = country["Total"]
    gold_pct = (country["Gold"] / total) * 100
    silver_pct = (country["Silver"] / total) * 100
    bronze_pct = (country["Bronze"] / total) * 100
    rank = country["Rank"]

    if gold_pct > 60:
        category = "Gold Dominant Category"
    elif silver_pct > 60:
        category = "Silver Dominant Category"
    elif bronze_pct > 60:
        category = "Bronze Dominant Category"
    elif abs(gold_pct - silver_pct) < 10 and abs(silver_pct - bronze_pct) < 10:
        category = "Balanced Category"
    else:
        category = "Mixed Category"

    Countries[category].append({
        "Country": country["Country"],
        "Total": total,
        "Gold %": gold_pct,
        "Silver %": silver_pct,
        "Bronze %": bronze_pct,
        "Rank": rank
    })

```

## Dense Ranking Unsorted by Total Implementation

```python

def assign_dense_ranks_unsorted_but_by_total(countries_dict):
    for category, countries in countries_dict.items():
        # Step 1: Make a sorted copy to get ranking order
        sorted_by_total = sorted(countries, key=lambda x: x["Total"], reverse=True)

        # Step 2: Build a Total → Rank mapping
        rank_map = {}
        current_rank = 1
        last_total = None

        for country in sorted_by_total:
            total = country["Total"]
            if total != last_total:
                rank_map[total] = current_rank
                current_rank += 1
                last_total = total

        # Step 3: Assign ranks back in original order
        for country in countries:
            country["Total Rank"] = rank_map[country["Total"]]


# Assign total ranks
assign_dense_ranks_unsorted_but_by_total(Countries)



# Display results
for category, countries in Countries.items():
    print(f"\n{category} (Without Sorting By Total)")
    for c in countries:
        print(f"{c['Country']} → Total: {c['Total']} medals, Gold: {c['Gold %']:.2f}%, Silver: {c['Silver %']:.2f}%, Bronze: {c['Bronze %']:.2f}%, Rank: {c['Total Rank']}")
```

## Sorting Total Medals Using Bubble Sort

```python
# Bubble Sort
def bubble_sort_by_total(data):
    arr = data.copy()
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j]["Total"] < arr[j + 1]["Total"]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


# Dense Ranking Function
def assign_dense_ranks(countries_dict, sort_func):
    for category, countries in countries_dict.items():
        sorted_list = sort_func(countries)
        current_rank = 1
        last_total = None
        for i, country in enumerate(sorted_list):
            if country["Total"] != last_total:
                country["Category Rank"] = current_rank
                last_total = country["Total"]
                current_rank += 1
            else:
                country["Category Rank"] = sorted_list[i - 1]["Category Rank"]
        countries_dict[category] = sorted_list
```

```python
assign_dense_ranks(Countries, bubble_sort_by_total)
```

```python
for category, countries in Countries.items():
    print(f"\n{category} (Sorting by Total Using Bubble Sort): ")
    for c in countries:
        print(f"{c['Country']} → Total: {c['Total']} medals, "
              f"Gold: {c['Gold %']:.2f}%, Silver: {c['Silver %']:.2f}%, Bronze: {c['Bronze %']:.2f}%, "
              f"Rank: {c['Category Rank']}")
```

## Sorting Total Medals Using Selection Sort

```python
# Selection Sort
def selection_sort_by_total(data):
    arr = data.copy()
    n = len(arr)
    for i in range(n):
        max_idx = i
        for j in range(i + 1, n):
            if arr[j]["Total"] > arr[max_idx]["Total"]:
                max_idx = j
        arr[i], arr[max_idx] = arr[max_idx], arr[i]
    return arr
```

```python
assign_dense_ranks(Countries, selection_sort_by_total)
```

```python
for category, countries in Countries.items():
    print(f"\n{category} (Sorting by Total Using Selection Sort):")
    for c in countries:
        print(f"{c['Country']} → Total: {c['Total']} medals, "
              f"Gold: {c['Gold %']:.2f}%, Silver: {c['Silver %']:.2f}%, Bronze: {c['Bronze %']:.2f}%, "
              f"Rank: {c['Category Rank']}")
```

## Sorting Total Medals Using Insertion Sort

```python
# Insertion Sort
def insertion_sort_by_total(data):
    arr = data.copy()
    for i in range(1, len(arr)):
        key_item = arr[i]
        j = i - 1
        while j >= 0 and arr[j]["Total"] < key_item["Total"]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key_item
    return arr



# Dense Ranking Function
def assign_dense_ranks(countries_dict, sort_func):
    for category, countries in countries_dict.items():
        sorted_list = sort_func(countries)
        current_rank = 1
        last_total = None
        for i, country in enumerate(sorted_list):
            if country["Total"] != last_total:
                country["Category Rank"] = current_rank
                last_total = country["Total"]
                current_rank += 1
            else:
                country["Category Rank"] = sorted_list[i - 1]["Category Rank"]
        countries_dict[category] = sorted_list

```

```python
assign_dense_ranks(Countries, insertion_sort_by_total)
```

```python
# Display Results
for category, countries in Countries.items():
    print(f"\n{category} (Sorting by Total Using Insertion Sort):")
    for c in countries:
        print(f"{c['Country']} → Total: {c['Total']} medals, "
              f"Gold: {c['Gold %']:.2f}%, Silver: {c['Silver %']:.2f}%, Bronze: {c['Bronze %']:.2f}%, "
              f"Rank: {c['Category Rank']}")
```

## Checking Time Efficiency of Insertion, Merge and Tim Sort Algorithms in Mixed Category

```python
import time

print("\nExecution Time for Sorting Algorithms on Mixed Category Countries:\n")

# Prepare sample
mixed_sample = Countries["Mixed Category"][:]

# Bubble Sort Timing
start_time = time.time()
bubble_sorted = bubble_sort_by_total(mixed_sample)
end_time = time.time()
bubble_time = end_time - start_time
print(f"Bubble Sort Time:    {bubble_time:.6f} seconds")


# Selection Sort Timing
start_time = time.time()
selection_sorted = selection_sort_by_total(mixed_sample)
end_time = time.time()
selection_time = end_time - start_time
print(f"Selection Sort Time: {selection_time:.6f} seconds")

# Insertion Sort Timing
start_time = time.time()
insertion_sorted = insertion_sort_by_total(mixed_sample)
end_time = time.time()
insertion_time = end_time - start_time
print(f"Insertion Sort Time: {insertion_time:.6f} seconds")


```
